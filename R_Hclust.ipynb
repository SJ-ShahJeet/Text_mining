```r
install.packages("proxy")
install.packages("cluster")
library(proxy)
library(stats)
```

```r
tfidf <- read.csv("tfidf_for_hclust.csv")
```

```r
cosine_dist <- dist(tfidf, method = "cosine")
```

```r
hclust_result <- hclust(cosine_dist, method = "ward.D2")
```

```r
plot(hclust_result, labels = FALSE, hang = -1, main = "Hierarchical Clustering Dendrogram (Cosine Distance)", xlab = "", sub = "")
rect.hclust(hclust_result, k = 4, border = "blue")
```
